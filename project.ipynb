```python
import pandas as pd
import zipfile

# Specify the path to the ZIP file
zip_file_path = "C:/Users/ana-m/Fundamental_data_2023/iris.zip"

# Extract the contents of the ZIP file to a temporary directory
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall("C:/Users/ana-m/Fundamental_data_2023/")

# Specify the path to the extracted CSV file
csv_file_path = "C:/Users/ana-m/Fundamental_data_2023/iris.data"

# Load the dataset
iris_data = pd.read_csv(csv_file_path, header=None, names=['sepal length', 'sepal width', 'petal length', 'petal width', 'species'])
```

```python
# Example for Sepal Length
sepal_length = iris_data['sepal length']
variable_type = 'Numeric' 
scale_of_measurement = 'Interval'

```

```python
# Example for Sepal Length
mean_sepal_length = sepal_length.mean()
std_sepal_length = sepal_length.std()
median_sepal_length = sepal_length.median()
```

```python
# Pair Plot for multivariate analysis
sns.pairplot(iris_data, hue="species", markers=["o", "s", "D"])
plt.suptitle("Pair Plot for Iris Dataset")
plt.show()
```

The pair plot provides a visual overview of the relationships between pairs of variables in the dataset. Each scatter plot in the grid represents the relationship between two variables, and the diagonal shows the distribution of each variable. In this case, different Iris species are distinguished by colors, and various markers are used for better clarity. This plot helps us observe potential patterns or clusters in the multivariate space.

```python
# Boxplot for Sepal Length
plt.figure(figsize=(8, 5))
sns.boxplot(x=iris_data.iloc[:, 4], y=iris_data.iloc[:, 0])
plt.title("Boxplot for Sepal Length Across Iris Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

The boxplot illustrates the distribution of a single variable across different categories (Iris species in this case). It displays key statistical measures such as median, quartiles, and potential outliers. This specific boxplot visualizes the distribution of sepal lengths for each Iris species, allowing us to compare the central tendency and spread of the data.

```python
# Boxplot for Sepal Length
plt.figure(figsize=(8, 5))
sns.boxplot(x=iris_data.iloc[:, 4], y=iris_data.iloc[:, 0])
plt.title("Boxplot for Sepal Length Across Iris Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

The violin plot is similar to a boxplot but also includes a kernel density estimation on each side. This plot provides insights into the distribution of petal widths for each Iris species. It combines aspects of both boxplot and KDE, offering a more comprehensive view of the data distribution, particularly in terms of density.

```python
# Exclude non-numeric columns for correlation analysis
numeric_columns = iris_data.select_dtypes(include='number')

# Create the correlation matrix
correlation_matrix = numeric_columns.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap for Iris Dataset")
plt.show()

```

The correlation heatmap showcases the pairwise correlation coefficients between numerical variables in the dataset. A color gradient is used to indicate the strength and direction of the correlation. This plot is useful for identifying relationships between variables. In the context of the Iris dataset, it helps us understand how different features correlate with each other.

```python
# Swarm Plot for Sepal Width
plt.figure(figsize=(10, 6))
sns.swarmplot(x=iris_data.iloc[:, 4], y=iris_data.iloc[:, 1])
plt.title("Swarm Plot for Sepal Width Across Iris Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

The swarm plot displays individual data points along an axis, avoiding overlap. In this case, it shows the distribution of sepal widths for each Iris species. The swarm plot provides a detailed view of the data distribution, allowing us to observe the density of data points and potential patterns within each category.
